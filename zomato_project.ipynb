```python
"""def rate_normalise(value):
    value=str(value).split('/')
    value=float(value[0])
    vlaue=int(value)
    return value"""

def rate_normalise(value):
    value = str(value).split('/')[0]
    return int(float(value))
```

```python
data_frame['rate']=data_frame['rate'].apply(rate_normalise)
```
